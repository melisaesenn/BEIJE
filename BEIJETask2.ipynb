```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import plotly

import chart_studio.plotly as py
import warnings
import plotly.graph_objs as go 
```

```python
dataset = pd.read_excel("Datasets\BEIJE _ Data Analytics & BI Intern Assignment - Dataset1.xlsx",
                       parse_dates = True)
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset.dropna(axis = "index", inplace=True)
```

```python
dataset.info()
```

```python
dataset['Date'] = pd.to_datetime(dataset['Date'], errors='ignore')
dataset.info()
```

```python
from prophet import Prophet
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8), dpi=100)
plt.plot(dataset['Date'], dataset['Number of Sales'])
plt.title('Pageview')
plt.xlabel('Date')
plt.ylabel('Number of Sales')
plt.show()
```

```python
m = Prophet()
```

```python
numberofsales = dataset[["Date", "Number of Sales"]]
```

```python
numberofsales = numberofsales.rename({"Date":"ds", "Number of Sales":"y"}, axis=1)
```

```python
numberofsales
```

```python
m.fit(numberofsales)
```

```python
future = m.make_future_dataframe(periods=4, freq='M', include_history=True)
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```
